```python
plt.figure(figsize=[7,7], dpi=150, facecolor='grey')
plt.subplot(111)
plt.plot(donnees['Height'], donnees['Weight'],
        marker='3', color='y', linestyle='', markersize=1.8)

plt.title('Poids en fonction de la raille des athltes des JO')
plt.ylabel('Poids')
plt.xlabel('Taille')
plt.show()
```

```python
plt.figure(figsize=[7,7])
plt.subplot(111)

plt.plot(donnees['Height'], donnees['Weight'],
        marker='3', color='g', linestyle='', markersize=2.5, label='Poids_vs_Taille')
plt.plot(donnees['Height'], donnees['Age'],
        marker='3', color='y', linestyle='', markersize=2.5, label='Age_vs_Taille')

plt.title('Athltes des JO')
plt.ylabel('Poids/Age')
plt.xlabel('Taille')

plt.legend(loc='upper left')
plt.show()
```

```python
plt.figure(figsize=[7,7])
plt.subplot(111)

plt.plot(donnees['Height'], donnees['Weight'],
        marker='3', color='g', linestyle='', markersize=2.5, label='Poids_vs_Taille')
plt.plot(donnees['Height'], donnees['Age'],
        marker='3', color='y', linestyle='', markersize=2.5, label='Age_vs_Taille')

plt.annotate('Moyene = 25 ans', xy=(210,25), xytext=(250,200), arrowprops=dict(facecolor='black'))

plt.title('Athltes des JO')
plt.ylabel('Poids/Age')
plt.xlabel('Taille')

plt.legend(loc='upper left')
plt.show()
```
