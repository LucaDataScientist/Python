```python
import pandas as pd

data = pd.read_excel (r'/home/simplon/Documents/Cours hebdomadaire/Semaine 4/Python/chomage_zone.xls', sheet_name=1, skiprows=[0,1,2,3,4],header=0)
print(data.head)
```

```python
# En utilisant groupby() et mean() successivement calculer le taux de chômage moyen par régions pour chaque trimestres. 
# Dans le groupby mettre as_index=‘False’. L’assigner à data_grp.

data_grp = data.groupby('LIBREG', as_index = False).mean()

print(data_grp)
```

```python
# Utiliser pandas.plot() pour tracer le diagramme à barre du taux de chômage du premier trimestre 2020. 
# Les noms de régions doivent apparaitre en abscisse.

# En vous appuyant sur la documentation en ligne donné une taille correct au graphique, lui donner un titre et ajouter une grille en arrière-plan.

data_grp.plot(x='LIBREG', y='2020-T1', kind='bar', figsize=(15,10), title='Taux de chomage du premier trimestre 2020', grid=True)
```

```python
# En repartant du jeux de données originel, isoler les colonnes 3 à 73. et assignées les à data2


data2 = data.iloc[:, 3:74] 

print(data2)
```

```python
# En utilisant groupby() et mean() successivement calculer le taux de chômage moyen par régions pour chaque trimestres. 
# Cette fois ci ne pas préciser l’instruction as_index. L’assigner à data_grp2.

data_grp2 = data2.groupby('LIBREG').mean()

print(data_grp2)
```

```python
# Transposer le dataframe

data_transpose = data_grp2.transpose()
data_transpose
```

```python
# Réindexer le dataframe avec reset_index()

data_index = data_transpose.reset_index()
data_index
```

```python
# Renommer le nom de la colonne ‘index’ en ‘periode’

data_index.rename(columns={'index': 'periode'}, inplace=True)

data_index
```

```python
# Après avoir visualisé le contenu de data2, utiliser pandas.plot(Provence-Alpes-Côte d'Azur) pour tracer la ligne
# représentant l’évolution du taux de chômage pour la région’

data_index.plot(x='periode', y="Provence-Alpes-Côte d'Azur", kind='line', figsize=(15,10), title='Taux de chomage pour la région Provence-Alpes-Cote dAzur', grid=True)
```

```python
# Répéter l’étape précédente sans préciser l’argument « y = »

data_index.plot(x='periode', kind='line', figsize=(15,10), title='Taux de chomage pour la région Provence-Alpes-Cote dAzur', grid=True)
```

```python
# Représenter l’évolution du taux de chômage pour la région ‘Provence-Alpes-Côte d'Azur’, 
# ‘Occitanie’ et ‘Ile-de-France’ sur un même graphique

data_index.plot(x='periode', y=["Provence-Alpes-Côte d'Azur", 'Occitanie', 'Ile-de-France'], kind='line', figsize=(15,10), title='Taux de chomage pour la région Provence-Alpes-Cote dAzur', grid=True)
```
